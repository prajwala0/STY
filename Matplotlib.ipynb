### import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### version 

```python
import matplotlib
matplotlib.__version__
```

### plot x-y graph

```python
x = np.array([0,1,2,3,4])
y = np.array([0,1,4,9,16])
plt.plot(x,y,'r')
plt.show
```

### using linspace

```python
x = np.linspace(0,10,6)
y = x**2
plt.plot(x,y)
plt.show()
```

### figure size

```python
plt.figure(figsize=(10,4))
plt.plot(x,y)
plt.show()
```

### different features

```python
plt.plot(x,y, color='hotpink', linestyle=':', linewidth='3')
plt.show()
```

```python
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])
plt.plot(y1, 'o')
plt.plot(y2, '*')
plt.show()
```

### subplots

```python
x = np.linspace(0,10,6)
y = x**2

plt.subplot(2,2,1)
plt.plot(x,y,'r*')

plt.subplot(2,2,2)
plt.plot(y,x,'yo')

plt.subplot(2,2,3)
plt.plot(x,y,'g')

plt.subplot(2,2,4)
plt.plot(y,x,'b')

plt.show()
```

### graph of inner graph

```python
x = np.linspace(0,5,11)
y = x**2

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.3,0.4])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

### subplots

```python
fig = plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=3)

plt.subplot(2,3,1)
plt.plot(x, y, 'r+')

plt.subplot(2,3,2)
plt.plot(y, x, 'y>')

plt.subplot(2,3,3)
plt.plot(x, y, 'r<')

plt.subplot(2,3,4)
plt.plot(y,x,'y*')
```

### maximum properties in graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=1, ls='-', marker='o', markersize=20, 
        markerfacecolor = 'yellow', markeredgewidth=3, markeredgecolor = 'green')
```
