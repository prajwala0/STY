### import library

```python
import seaborn as sns
```

### get list of inbuilt datasets from seaborn

```python
sns.get_dataset_names()
```

### load dataset

```python
tips = sns.load_dataset('tips')
tips
```

```python
flights = sns.load_dataset('flights')
flights
```

### distplot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=5, label='total bills' )
```

### histplot

```python
sns.histplot(tips['total_bill'],kde=True)
```

### jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

### pairplot 

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

### rugplot

```python
sns.rugplot(tips['total_bill'])
```

A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis.

```python
sns.rugplot(x=flights['year'], y=flights['passengers'])
```

### barplot

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

### countplot

```python
sns.countplot(x='sex', data =tips)
```

### boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

### violinplot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

### stripplot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter =True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter =0.5)
```

jitter can be used to provide displacements along the horizontal axis, which is useful when there are large clusters of data points. You can specify the amount of jitter (half the width of the uniform random variable support), or just use True for a good default.

### swarmplot

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

### relplot

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", col="time")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", col="time", row="sex")
```

### lineplot

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

### kdeplot

```python
sns.kdeplot(data=tips, x="total_bill", bw_adjust=.2)
```

```python
sns.kdeplot(data=tips, x="total_bill", bw_adjust=2, hue="time")
```

### catplot

```python
sns.catplot(x='day', y='total_bill', data=tips, hue="sex", kind="boxen")
```

```python
sns.catplot(x='day', y='total_bill', data=tips, hue="sex", kind="violin")
```

### regplot

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

### heatmap

```python
import matplotlib.pyplot as plt
flights = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(12,6))
sns.heatmap(flights)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(flights, annot=True, fmt=".1f")
```
